```python
import numpy as np
import pandas as pd
# Load in csv
vine_table_df = pd.read_csv("vine_table.csv")
vine_table_df
```

```python
vote_df = vine_table_df[vine_table_df["total_votes"] >= 20]
vote_df.head()
```

```python
helpful_df = vote_df.loc[vote_df.helpful_votes / vote_df.total_votes >= .5]
helpful_df
```

```python
# Vine program
vine_df = helpful_df[helpful_df["vine"] == "Y"]
vine_df
```

```python
# Not vine program
novine_df = helpful_df[helpful_df["vine"] == "N"]
novine_df
```

```python
# Total number vine reviews
len(vine_df)
```

```python
# Total number not vine reviews
len(novine_df)
```

```python
# Total number vine 5 star reviews
len(vine_df[vine_df["star_rating"]==5])
```

```python
# Total number not vine 5 star reviews
len(novine_df[novine_df["star_rating"]==5])
```

```python
# Total number helpful all with more than 20 votes over 50 percent positive votes
len(helpful_df)
```
